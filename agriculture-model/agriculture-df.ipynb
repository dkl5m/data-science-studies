```python
# import required packages
import numpy as np
import pandas as pd
```

Loading dataset using pandas

```python
#@title
df=pd.read_csv('agricultural_raw_material.csv')
```

Exploring dataset

```python
df.info
#Checking Null Values of each column
df.isnull().sum()
```

Treating missing, incorrect and invalid data

```python
# Converting to lowercase letters
df.columns = [str(i).lower() for i in df.columns]
# Replacing %, "," and "-"
df = df.replace('%', '', regex=True)
df = df.replace(',', '', regex=True)
df = df.replace('-', '', regex=True)
df = df.replace('', np.nan)
df = df.replace('MAY90', np.nan)
# Dropping rows with NaN values
df = df.dropna()
# Check to see if all NaN vlaues are resolved
df.isnull().sum()
# Converting data type to float
lst = df.columns[1:]
df[lst] = df[lst].astype('float')
df.dtypes
```

```python
df.head()
```

Date and hours columns

```python
#df['Month'] = df['Month'].astype("string")
df.month = pd.to_datetime(df.month.str.upper(), format='%b%y', yearfirst=False)
# Indexing month
df = df.set_index('month')

df.head()
```

Exploratory analysis and visualization

```python
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
sns.set_style('darkgrid')
matplotlib.rcParams['font.size'] = 14
matplotlib.rcParams['figure.figsize'] = (9, 5)
matplotlib.rcParams['figure.facecolor'] = '#00000000'
```

Heat Map

```python
#raw-materials list
raw_data=['coarse wool price', 'copra price', 'cotton price', 'fine wool price', 'hard log price', 'hard sawnwood price',
 'hide price', 'plywood price', 'rubber price', 'softlog price', 'soft sawnwood price', 'wood pulp price']
# getting the correlation matrix
corrmat = df[raw_data].corr()
# setting the size of the plot
fig = plt.figure(figsize = (12, 9))
# masking the upper triangle part since matrix is symmetric (repetitive)
mask = np.triu(np.ones_like(corrmat, dtype=bool))
sns.heatmap(corrmat, vmax = .8, mask = mask, square = True, annot = True)
plt.show()
```

        It shows that has some correlation between the prices of some raw materials

```python
plt.figure(figsize=(30,15))
changeList=['coarse wool price % change', 'copra price % change', 'cotton price % change', 'fine wool price % change', 'hard log price % change', 'hard sawnwood price % change',
 'hide price % change', 'plywood price % change', 'rubber price % change', 'softlog price % change', 'soft sawnwood price % change', 'wood pulp price % change']
# generate a correlation matrix for the whole dataset
corrMatrix = df[changeList].corr()
sns.heatmap(corrMatrix, annot=True)
plt.show()
```

        it shows that there's almost no correlation between price % change (strong correlation > 0.7)
